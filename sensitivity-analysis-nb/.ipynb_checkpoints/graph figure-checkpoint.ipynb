```python
import networkx as nx
import matplotlib.pyplot as plt

# Initialize an empty graph
G = nx.Graph()

# Path to your DIMACS file
file_path = 'myciel3.txt'

# Read the DIMACS file
with open(file_path, 'r') as file:
    lines = file.readlines()

# First line contains the number of nodes and edges (not strictly needed)
num_nodes, num_edges = map(int, lines[0].split())

# Add edges to the graph
for line in lines[1:]:
    u, v = map(int, line.split())
    G.add_edge(u, v)

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', font_weight='bold', node_size=700)
plt.title("Graph Visualization from DIMACS file")
plt.show()
```

```python
def random_int(n):      #PENSO SIA SUPERFLUA! posso usare direttamente np.random.randint dopo nel codice
    """Randomly generate an integer from 0 to (n-1)."""
    new_var = np.random.randint
    return new_var(n) 
```
